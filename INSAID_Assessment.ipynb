```python
train, test = train_test_split(df, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
train_X = train[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','nameDest','oldbalanceDest','newbalanceDest','isFraud','isFlaggedFraud']]
train_y = train.isFraud

test_X = test[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','nameDest','oldbalanceDest','newbalanceDest','isFraud','isFlaggedFraud']]
test_y = test.isFraud
```

```python
train_X = train[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','nameDest','oldbalanceDest','newbalanceDest','isFraud','isFlaggedFraud']]
train_y = train.isFlaggedFraud

test_X = test[['amount','nameOrig','oldbalanceOrg','newbalanceOrig','nameDest','oldbalanceDest','newbalanceDest','isFraud','isFlaggedFraud']]
test_y = test.isFlaggedFraud
```

```python
#SVM
model = svm.SVC()
model.fit(train_X, train_y)
prediction = model.predict(test_X)
print("The Accuracy of SVM is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#KNN
model1 = KNeighborsClassifier(n_neighbors = 3)
model1.fit(train_X, train_y)
prediction = model1.predict(test_X)
print("The Accuracy of KNN is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#Logistic Regression
model2  = LogisticRegression()
model2.fit(train_X, train_y)
prediction = model2.predict(test_X)
print("The Accuracy of LR is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#Decision Tree - CART
model3 = DecisionTreeClassifier()
model3.fit(train_X, train_y)
prediction = model3.predict(test_X)
print("The Accuracy of DT is: ",metrics.accuracy_score(prediction, test_y))
```
