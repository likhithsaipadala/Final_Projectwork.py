```python
train, test = train_test_split(cancer, test_size = 0.3)
print(train.shape)
print(test.shape)
```

```python
train_X = train[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave_points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst' ]]
train_y = train.diagnosis

test_X = test[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave_points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst' ]]
test_y = test.diagnosis
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
#SVM
model = svm.SVC()
model.fit(train_X, train_y)
prediction = model.predict(test_X)
print("The Accuracy of SVM is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#KNN
model1 = KNeighborsClassifier(n_neighbors = 3)
model1.fit(train_X, train_y)
prediction = model1.predict(test_X)
print("The Accuracy of KNN is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#Logistic Regression
model2  = LogisticRegression()
model2.fit(train_X, train_y)
prediction = model2.predict(test_X)
print("The Accuracy of LR is: ",metrics.accuracy_score(prediction, test_y))
```

```python
#Decision Tree - CART
model3 = DecisionTreeClassifier()
model3.fit(train_X, train_y)
prediction = model3.predict(test_X)
print("The Accuracy of DT is: ",metrics.accuracy_score(prediction, test_y))
```

```python
mean = cancer[['radius_mean', 'texture_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean','diagnosis']]
standard_error = cancer[['radius_se','texture_se', 'perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave_points_se','symmetry_se','fractal_dimension_se','diagnosis']]
worst = cancer[['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst','diagnosis']]
```

```python
#For Mean

train_M, test_M = train_test_split(mean, test_size = 0.3, random_state =0)
train_x_M = train_M[['radius_mean', 'texture_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean' ]]
train_y_M = train_M.diagnosis

test_x_M = test_M[['radius_mean', 'texture_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean','fractal_dimension_mean' ]]
test_y_M = test_M.diagnosis
```

```python
#For standard_error

train_SE, test_SE = train_test_split(standard_error, test_size = 0.3, random_state =0)
train_x_SE = train_SE[['radius_se','texture_se', 'perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave_points_se','symmetry_se','fractal_dimension_se']]
train_y_SE = train_SE.diagnosis

test_x_SE = test_SE[['radius_se','texture_se', 'perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave_points_se','symmetry_se','fractal_dimension_se']]
test_y_SE = test_SE.diagnosis
```

```python
#For Worst

train_W, test_W = train_test_split(worst, test_size = 0.3, random_state =0)
train_x_W = train_W[['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst']]
train_y_W = train_W.diagnosis

test_x_W = test_W[['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst']]
test_y_W = test_W.diagnosis
```

```python
# 1. SVM
model = svm.SVC()
model.fit(train_x_M,train_y_M)
prediction = model.predict(test_x_M)
print("The Accuracy of mean is: ",metrics.accuracy_score(prediction, test_y_M))

model1 = svm.SVC()
model1.fit(train_x_SE, train_y_SE)
prediction1 = model1.predict(test_x_SE)
print("The Accuracy of Standard_Error is: ",metrics.accuracy_score(prediction1, test_y_SE))


model2 = svm.SVC()
model2.fit(train_x_W, train_y_W)
prediction2 = model2.predict(test_x_W)
print("The Accuracy of Worst is: ",metrics.accuracy_score(prediction2, test_y_W))
```

```python
#Logistic_Regression
model3  = LogisticRegression()
model3.fit(train_x_M, train_y_M)
prediction3 = model3.predict(test_x_M)
print("The Accuracy of LR for mean is: ",metrics.accuracy_score(prediction3, test_y_M))

model4  = LogisticRegression()
model4.fit(train_x_SE, train_y_SE)
prediction4 = model4.predict(test_x_SE)
print("The Accuracy of LR for standard Error is: ",metrics.accuracy_score(prediction4, test_y_SE))

model5  = LogisticRegression()
model5.fit(train_x_W, train_y_W)
prediction5 = model5.predict(test_x_W)
print("The Accuracy of LR for Worst is: ",metrics.accuracy_score(prediction5, test_y_W))

```

```python
#Desicion Tree
model6 = DecisionTreeClassifier()
model6.fit(train_x_M, train_y_M)
prediction6 = model6.predict(test_x_M)
print("The Accuracy of mean is: ",metrics.accuracy_score(prediction6, test_y_M))

model7 = DecisionTreeClassifier()
model7.fit(train_x_SE, train_y_SE)
prediction7 = model7.predict(test_x_SE)
print("The Accuracy of Standard Error is: ",metrics.accuracy_score(prediction7, test_y_M))

model8 = DecisionTreeClassifier()
model8.fit(train_x_W, train_y_W)
prediction8 = model8.predict(test_x_W)
print("The Accuracy of mean is: ",metrics.accuracy_score(prediction8, test_y_W))

```

```python
#KNN

model9 = KNeighborsClassifier(n_neighbors = 3)
model9.fit(train_x_M, train_y_M)
prediction9 = model9.predict(test_x_M)
print("The Accuracy of KNN mean is: ",metrics.accuracy_score(prediction9, test_y_M))

model10 = KNeighborsClassifier(n_neighbors = 3)
model10.fit(train_x_SE, train_y_SE)
prediction10 = model10.predict(test_x_SE)
print("The Accuracy of KNN standard error is: ",metrics.accuracy_score(prediction10, test_y_SE))

model11 = KNeighborsClassifier(n_neighbors = 3)
model11.fit(train_x_W, train_y_W)
prediction11 = model11.predict(test_x_W)
print("The Accuracy of KNN worst is: ",metrics.accuracy_score(prediction10, test_y_W))
```
